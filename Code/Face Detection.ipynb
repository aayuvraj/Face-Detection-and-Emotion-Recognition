```python
import numpy as np
from keras.utils import to_categorical
import os
import glob
import cv2 
import math
import dlib
import tensorflow as tf
from scipy.linalg import svd
from numpy import linalg as LA

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split



%matplotlib inline
```

```python
path = glob.glob("F:/Program Files/Final_Dataset/*RGB.png")
```

```python
path_r="F:/Program Files/Final_Dataset/"
```

```python
face_write_path = 'F:/Program Files/Detected'
```

```python
B=[]
for f in os.listdir(path_r):
    if"RGB" in f:
        print(f)
        B.append(f)
    
```

```python
len(B)
```

```python
X=[]
hogFaceDetector = dlib.get_frontal_face_detector()
width =48
height =48
cnt=0
for img in path:
    im = cv2.imread(img)
    faceRects = hogFaceDetector(im, 0)
    for faceRect in faceRects:
        x = faceRect.left()
        y = faceRect.top()
        x2 = faceRect.right()
        y2 = faceRect.bottom()
    c_img = im[y:y2,x:x2]
    gray = cv2.cvtColor(c_img,cv2.COLOR_BGR2GRAY)
    gray = cv2.resize(gray,(height,width))
    X.append(gray)
    filename=B[cnt]
    cv2.imwrite(os.path.join(face_write_path ,filename),gray)
    cnt=cnt+1
    
```
